```python
import re

def is_valid_email(email):
    pattern = r'^[A-Za-z][A-Za-z0-9._-]*@[A-Za-z0-9.-]+\.[A-Za-z]{2,}$'
    return re.match(pattern, email) is not None

def main():
    emails = [
        "alice@example.com",
        "bob_123@gmail.com",
        "1invalid@gmail.com",
        "charlie.test@company.com"
    ]

    for email in emails:
        print(email, "->", is_valid_email(email))

if __name__ == "__main__":
    main()
```
