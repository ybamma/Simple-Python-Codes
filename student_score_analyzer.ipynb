```python
def calculate_average(scores):
    return sum(scores) / len(scores) if scores else 0

def get_grade(avg):
    if avg >= 90:
        return "A"
    elif avg >= 80:
        return "B"
    elif avg >= 70:
        return "C"
    elif avg >= 60:
        return "D"
    else:
        return "F"

def main():
    students = {
        "Aarav": [85, 90, 88],
        "Sophia": [92, 95, 94],
        "Liam": [70, 75, 72],
        "Emma": [60, 58, 65]
    }

    print("Student Score Report\n")

    for name, scores in students.items():
        avg = calculate_average(scores)
        grade = get_grade(avg)
        print(f"{name}: Average = {avg:.2f}, Grade = {grade}")

if __name__ == "__main__":
    main()
```
